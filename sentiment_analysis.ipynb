```python
import pandas as pd
data_columns=['Id','Location','target','text']
df=pd.read_csv('/content/twitter_validation.csv',names=data_columns)
df
```

```python
df.isna().sum()
```

```python
df['target'].value_counts()
```

```python
df['target'].unique() # to find how many unique values
```

```python
# to plot the count, use countplot
import seaborn as sns
sns.countplot(x='target',order=df['target'].value_counts(ascending=True).index,data=df)
```

```python
# drop target=irrelevant, row wise
df.drop(df.index[(df['target']=='Irrelevant')],axis=0,inplace=True)
df
```

```python
#reset index, now inbetween rows are removed, drop= true used to remove the already existing index,have created new index
df.reset_index(drop=True,inplace=True)
df
```

```python
# drop location and id
df.drop(['Id','Location'],axis=1,inplace=True)
df
```

```python
# assign values to positive,negative and neutral
df['target']=df['target'].map({'Positive':1,'Negative':-1,'Neutral':0})
df
```

```python
# take text only
tweets=df.text
tweets
```

```python
# remove special characters
tweets=tweets.str.replace('[^a-zA-Z0-9]+',' ')
tweets
```

```python
from nltk.stem import SnowballStemmer
stemmer=SnowballStemmer('english')
from nltk import TweetTokenizer
tk=TweetTokenizer()
tweets=tweets.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda x:' '.join(x))
```

```python
tweets
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stop=stopwords.words('english')
tweets=tweets.apply(lambda x:[i for i in tk.tokenize(x) if i not in stop]).apply(lambda x:' '.join(x))
tweets
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer()
train_data_vec=vec.fit_transform(tweets)
print(train_data_vec)   # this is x value      # 0 row, 739 column
```

```python
train_data_vec.shape # 3931 features or columns
```

```python
# convert target to numbers
y=df['target'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data_vec,y,test_size=0.3,random_state=1)
```

```python
from sklearn import svm
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
svm_model=svm.SVC()
NB_model=MultinomialNB()
RF_model=RandomForestClassifier()
AB_model=AdaBoostClassifier()
model=[svm_model,NB_model,RF_model,AB_model]
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,classification_report
for i in model:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('**************************************************')
  print(classification_report(y_test,y_pred))
  print('***************************************************')

```
